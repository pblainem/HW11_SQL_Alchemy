```python
#import csv files
import os
import pandas as pd
import csv
from sqlalchemy import create_engine
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String, Float
from sqlalchemy.orm import sessionmaker
from sqlalchemy.sql import *
from pandas.io.sql import SQLTable, SQLDatabase
from pprint import pprint
from datetime import datetime



cln_HImeasCSVpath = os.path.join("Resources", "clean_hawaii_measurements.csv")
cln_HIstationCSVpath= os.path.join("Resources", "clean_hawaii_stations.csv")

measurementsDF = pd.read_csv(cln_HImeasCSVpath)
stationsDF = pd.read_csv(cln_HIstationCSVpath)

measurementsDF.drop('Unnamed: 0', axis=1, inplace=True)
stationsDF.drop('Unnamed: 0', axis=1, inplace=True)

measurementsDF.head()
```

```python
# Use 3 slashes when connecting to sqlite database
path_to_file = os.path.join("Resources", "hawaii.sqlite")
engine = create_engine(f'sqlite:///{path_to_file}')
```

```python
Base = declarative_base()

class Measurement(Base):
    __tablename__ = 'measurements'
    id = Column(Integer, primary_key=True)
    station = Column(String(255))
    date = Column(String(255))
    prcp = Column(Float(2,2))
    tobs = Column(Integer)
    
class Station(Base):
    __tablename__ = 'stations'
    id = Column(Integer, primary_key=True)
    station = Column(String(255))
    name = Column(String(255))
    latitude = Column(Float(2,4))
    longitude = Column(Float(3,4))
    elevation = Column(Float(3,1))


Measurement.__table__.create(bind=engine, checkfirst=True)
Station.__table__.create(bind=engine, checkfirst=True)
    

    
```

```python
measurements, stations = [], []



for index, row in measurementsDF.iterrows():
#     print(row)
    row1 = {}
    row1['id'] = index
    row1['station'] = row['station']
    row1['prcp'] = row['prcp']
#     row1['date'] = datetime.strptime(row['date'], "%Y-%m-%d")
    row1['date'] = row['date']
    row1['tobs'] = row['tobs']
    
    measurements.append(row1)

for index, row in stationsDF.iterrows():
#     print(row)
    row1 = {}
    row1['id'] = index
    row1['station'] = row['station']
    row1['name'] = row['name']
    row1['latitude'] = row['latitude']
    row1['longitude'] = row['longitude']
    row1['elevation'] = row['elevation']
    
    stations.append(row1)    
    
    

pprint(stations)
```

```python
Session = sessionmaker(bind=engine)
session = Session()

for measurement in measurements:
    row = Measurement(**measurement)
    session.add(row)

for station in stations:
    row = Station(**station)
    session.add(row)

session.commit()
```
